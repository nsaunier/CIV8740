```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
class User:
    def __init__(self, x0, v0, sigma):
        self.x=x0
        self.v=v0
        self.sigma = sigma
        
    def update(self, a, deltat):
        self.v = self.v+a*deltat
        self.x = self.x+self.v*deltat+random.gauss(0.,self.sigma)
        return self.x, self.v
    
def update(user, a, nSteps):
    v = []
    x = []
    for i in range(nSteps):
        newx, newv = user.update(a,deltat)
        v.append(newv)
        x.append(newx)
    return x,v
```

```python
#seed = 123
#random.seed(123)
x0 = 0. # km
v0 = 0. # km/h
deltat = 0.1 # s

plt.figure(figsize=(14,8))
for i in range(10):
    x = [x0]
    v = [v0]
    car = User(x0/1000., v0*3.6, 0.1) # all in m and s
    x2,v2 = update(car, 0.5, 100)
    x = x+x2
    v = v+v2
    x2,v2  = update(car, 0., 100)
    x = x+x2
    v = v+v2
    plt.plot(x)
```
